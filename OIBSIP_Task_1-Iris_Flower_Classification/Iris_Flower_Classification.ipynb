# Specifying x and y 

```python
x = df.drop('Species', axis = 1)
x
```

```python
y = df["Species"]
y.head()
```

```python
y = pd.get_dummies(y)
# This is done to convert the categorical data/values in the y column into numerical data/values
```

```python
y
```

## Importing necessary libraries 

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score
```

# Split the dataset into training and testing sets

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42) 
```

```python
print(df.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Train the model using K-Nearest Neighbors Algorithm 

```python
# Initialize the K-Nearest Neighbors classifier
knn_classifier = KNeighborsClassifier(n_neighbors=3)
```

```python
# Train the classifier
knn_classifier.fit(x_train, y_train)
```

```python
# Make predictions on the testing set
y_pred = knn_classifier.predict(x_test)
```

```python
y_pred
```

# Checking Model Performance 

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
accuracy_percentage = accuracy * 100
print(f'Accuracy in Percentage: {accuracy_percentage:.2f}%')
```

```python
precision = precision_score(y_test, y_pred, average='weighted')
print(f'Precision: {precision}')
precision_percentage = precision * 100
print(f'Precision in Percentage: {precision_percentage:.2f}%')
```

```python
recall = recall_score(y_test, y_pred, average='weighted')
print(f'Recall: {recall}')
recall_percentage = recall * 100
print(f'Recall in Percentage: {recall_percentage:.2f}%')
```

```python
f1 = f1_score(y_test, y_pred, average='weighted')
print(f'F1 Score: {f1}')
f1score_percentage = f1 * 100
print(f'F1 Score in Percentage: {f1score_percentage:.2f}%')
```

```python
# Convert one-hot encoded labels to original class labels
y_true_labels = y_test.idxmax(axis=1)
y_pred_labels = pd.DataFrame(y_pred, columns=y_test.columns).idxmax(axis=1)

# Calculate the confusion matrix
conf_matrix = confusion_matrix(y_true_labels, y_pred_labels)
print('Confusion Matrix:\n', conf_matrix)
```

```python
classification_rep = classification_report(y_test, y_pred)
print('Classification Report:\n', classification_rep)
```

```python
# Heatmap for Model Evaluation
plt.figure(figsize=(10, 8))
ax = sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='GnBu', linewidths=.5, cbar=True,
                 xticklabels=y_test.columns, yticklabels=y_test.columns)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix for Iris Dataset Prediction', fontsize=16)
cbar = ax.collections[0].colorbar
cbar.set_label('Count', rotation=270, labelpad=15)
plt.show()
```

```python
new_features = pd.DataFrame({
    'SepalLengthCm': [5.0],
    'SepalWidthCm': [3.2],
    'PetalLengthCm': [1.2],
    'PetalWidthCm': [0.8]
})

# Make predictions using the trained model
predicted_labels = knn_classifier.predict(new_features)
predicted_class = y.columns[predicted_labels.argmax(axis=1)].values[0]
print('Predicted Iris Class:', predicted_class)
```
